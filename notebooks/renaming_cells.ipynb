```python
import pandas as pd
import os
import sys
import re
import numpy as np
```

```python
sample_regex = re.compile(r"GSM[0-9]{7}")

def get_sample_id(filename):
    return sample_regex.search(filename).group()
```

```python
df = pd.read_csv("data/GSM3478792_P5_matrix.csv")
df.head()
```

```python
sample_id = get_sample_id("data/GSM3478792_P5_matrix.csv")
sample_id
```

```python
df.shape
```

```python
number_of_cells = df.shape[1] - 1 # -1 za Index(gene) kolonu
```

```python
df.columns = [df.columns[0]] + [sample_id + "_" + str(cell_number) for cell_number in range(1, number_of_cells+1)]
df.head()
```

```python
df = df.set_index("Index")
df = df.transpose()
df.head()
```
