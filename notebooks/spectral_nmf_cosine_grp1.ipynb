```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import SpectralClustering
from sklearn.metrics import silhouette_score, silhouette_samples
import math
from sklearn.neighbors import NearestNeighbors, kneighbors_graph
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
df = pd.read_csv("grp1_nmf_50_dims.csv")
tsne_data = pd.read_csv("tsne5000_grp1.csv")
```

```python
tsne_data['cell_id'] = df['cell_id']
#df = df.drop(columns=['cell_id'])
```

```python
tsne_data.head()
```

```python
def scatter_plot_clusters(df):
    plt.figure(figsize=(8, 7))
    number_of_clusters = max(df['labels']) + 1
    for j in range(-1, number_of_clusters):
        if j == -1:
            label = 'noise'
        else:
            label = 'cluster %d' % j
            
        cluster = df.loc[df['labels'] == j]
        plt.scatter(x=cluster.iloc[:, 0], y=cluster.iloc[:, 1], label=label, alpha=0.75, edgecolor='black')
    plt.legend()
```

```python
est = SpectralClustering(n_clusters=5, affinity='precomputed').fit(kneighbors_graph(df, n_neighbors=10, mode='distance', metric='cosine'))
tsne_data['labels'] = est.labels_
scatter_plot_clusters(tsne_data)
```

```python
tsne_data['labels'] = est.labels_
```

```python
plt.figure(figsize=(8, 7))
cluster_0 = tsne_data.loc[tsne_data['labels'] == 0]
plt.scatter(x=tsne_data.iloc[:, 0], y=tsne_data.iloc[:, 1], edgecolor='black', label='Ostali podaci')
plt.scatter(x=cluster_0.iloc[:, 0], y=cluster_0.iloc[:, 1], edgecolor='black', label='Klaster od interesa')
plt.legend()
```

```python
file_distribution_per_cluster = cluster_0['cell_id'].apply(lambda cell_id: cell_id[:10]).value_counts()
print(file_distribution_per_cluster)
file_distribution_per_cluster.plot.bar(figsize=(6, 6), title='Raspodela fajlova u klasteru 0',color='#ff7f0e', grid=True, align='edge')
plt.xticks(rotation=45)
```
