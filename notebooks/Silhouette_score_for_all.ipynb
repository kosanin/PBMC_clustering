```python
import pandas as pd
import os
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv("data/version1_group1/grp1_nmf_50_dims.csv")
df = df.drop(columns=['cell_id'])
```

```python
labels = pd.read_csv("data/version1_group1/group1_clust_labels.csv")
labels.head()
```

```python
sil_score = []
for column in labels.columns[1:]:
    score = silhouette_score(X=df, labels=labels[column], metric='euclidean')
    print(column, score)
    sil_score.append((column, score))
```

```python
sil_score_cosine = [] 
for column in labels.columns[1:]:
    score = silhouette_score(X=df, labels=labels[column], metric='cosine')
    print(column, score)
    sil_score_cosine.append((column, score))
```

```python
sil_scores = pd.DataFrame()
```

```python
sil_scores['algorithm'] = list(map(lambda x: x[0], sil_score))
sil_scores['sil_coef_euclidean'] = list(map(lambda x: x[1], sil_score))
sil_scores['sil_coef_cosine'] = list(map(lambda x: x[1], sil_score_cosine))
sil_scores.head()
```

```python
sil_scores.to_csv("data/version1_group1/sil_scores.csv", index=False)
```
